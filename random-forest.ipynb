```python
X = df.drop(['Class'], axis = 1) 
Y = df["Class"] 
print(X.shape) 
print(Y.shape) 
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier 
rfc = RandomForestClassifier() 
rfc.fit(X_train, y_train) 
yPred = rfc.predict(X_test) 
```

```python
accuracy_rf = accuracy_score(y_test, yPred)
classification_report_rf = classification_report(y_test, yPred)

# Print best parameters and results
print(f"Random Forest Accuracy: {accuracy_rf}")
print("Random Forest Classification Report:")
print(classification_report_rf)
```

```python
from sklearn.svm import SVC
model2 = SVC(probability=True, random_state=2)
svm = model2.fit(X_train, y_train)
```

```python
y_pred = model2.predict(X_test)
svc_acc = accuracy_score(y_test, y_pred)
classification_report_svc = classification_report(y_test, y_pred)
```
